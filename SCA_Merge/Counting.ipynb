```python
def get_uniques(x):
    return list(x.unique())
```

```python
langlvl=l34.groupby('level_id').agg(
    n_anon_ids=('anon_id', 'nunique'),
    lang=('L1',"count")
)

#changes the clean_ids_lvl's index so we can access the anon_id easier
langlvl.reset_index(inplace=True)
langlvl
```

# Whole Dataset

```python
whole=whole.query("level_id!=2")
```

```python
df34=whole.query("level_id==3 or level_id==4")
df45=whole.query("level_id==4 or level_id==5")
```

```python
# this code gets how many times an id appeared (n_anon_ids) as well as an array of what levels they participated in (lvl_nums)
clean_ids=df34.groupby('anon_id').agg(
    n_anon_ids=('level_id', 'nunique'),
    lvl_nums=('level_id', get_uniques)
)

#changes the clean_ids_lvl's index so we can access the anon_id easier
clean_ids.reset_index()
clean_ids
```

```python
w34=clean_ids.query("n_anon_ids==2")
w34.reset_index(inplace=True)
```

```python
len(w34)
```

```python
x=w34['anon_id'].tolist()
```

```python
df34.reset_index(inplace=True)
```

```python
print(x)
```

```python
ind=0
for i in df34.anon_id:
    if i not in x:
        df34.drop(index=ind,inplace=True)
    ind+=1
df34.reset_index(inplace=True)
```

```python
df34=df34.drop(columns=['level_0','index','Unnamed: 0'])
```

```python
df34
```

```python
# this code gets how many times an id appeared (n_anon_ids) as well as an array of what levels they participated in (lvl_nums)
q_ids34=df34.groupby('anon_id').agg(
    n_anon_ids=('question_id', 'nunique'),
    lvl_nums=('level_id', get_uniques),
    q_nums=('question_id',get_uniques)
)

#changes the clean_ids_lvl's index so we can access the anon_id easier
q_ids34.reset_index(inplace=True)
q_ids34
```

```python
qgroup_ids34=df34.groupby('question_id').agg(
    n_anon_ids=('anon_id', 'nunique'),
    anon_ids=('anon_id',get_uniques),
    langs=("native_language","nunique")
)

#changes the clean_ids_lvl's index so we can access the anon_id easier
qgroup_ids34.reset_index(inplace=True)
qgroup_ids34
```

```python
whole34=qgroup_ids34.query("langs>2")
```

```python
len(whole34)
```
