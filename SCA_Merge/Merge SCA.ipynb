# Merging Whole dataset SCA analysis

```python
wchin34=pd.read_csv('C:\\Users\Hanna\Documents\GitHub\LRDC-Research\SCA_Analysis\wchin34.csv')
wchin45=pd.read_csv('C:\\Users\Hanna\Documents\GitHub\LRDC-Research\SCA_Analysis\wchin45.csv')
wchin35=pd.read_csv('C:\\Users\Hanna\Documents\GitHub\LRDC-Research\SCA_Analysis\wchin35.csv')

wspan34=pd.read_csv('C:\\Users\Hanna\Documents\GitHub\LRDC-Research\SCA_Analysis\wspan34.csv')
wspan45=pd.read_csv('C:\\Users\Hanna\Documents\GitHub\LRDC-Research\SCA_Analysis\wspan45.csv')
wspan35=pd.read_csv('C:\\Users\Hanna\Documents\GitHub\LRDC-Research\SCA_Analysis\wspan35.csv')

wkor34=pd.read_csv('C:\\Users\Hanna\Documents\GitHub\LRDC-Research\SCA_Analysis\wkor34.csv')
wkor45=pd.read_csv('C:\\Users\Hanna\Documents\GitHub\LRDC-Research\SCA_Analysis\wkor45.csv')
wkor35=pd.read_csv('C:\\Users\Hanna\Documents\GitHub\LRDC-Research\SCA_Analysis\wkor35.csv')
```

```python
#merging chinese data
wchin=wchin34.merge(wchin45,how='outer').merge(wchin35,how='outer')
wchin
```

```python
#merging spanish
wspan=wspan34.merge(wspan45,how='outer').merge(wspan35,how='outer')
wspan
```

```python
wspan.drop_duplicates(subset=['answer_id'],keep='last',inplace=True)
wspan=wspan.reset_index()
wspan.drop(columns=['Unnamed: 0','index'],inplace=True)
print(len(wspan))
wspan.head()
```

```python
#merging korean data
wkor=wkor34.merge(wkor45,how='outer').merge(wkor35,how='outer')
wkor
```

```python
wkor.drop_duplicates(subset=['answer_id'],keep='last',inplace=True)
wkor=wkor.reset_index()
wkor.drop(columns=['Unnamed: 0','index'],inplace=True)
print(len(wkor))
wkor.head()
```

```python
#merging all
SCAwhole=wchin.merge(wspan,how='outer').merge(wkor,how='outer')
SCAwhole
```

```python
#sending to csv file
SCAwhole.to_csv('C:\\Users\Hanna\Documents\GitHub\LRDC-Research\SCA_Merge/SCAwhole.csv')
```
