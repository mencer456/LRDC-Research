```python
import pandas as pd
```

```python
whole=pd.read_csv("CSV/whole_no_duplicates.csv")
loose=pd.read_csv("CSV/loose_no_duplicates.csv")
```

```python
whole.head(5)
```

```python
index=0
whole_txts=whole.copy()
whole_txts['fixed']=pd.NaT

for text in whole_txts.text:
    text=text.replace('. ','.').replace(', ',',') #removes extra space after . or ,
    s = text.replace(",",", ").replace(".",". ") #adds double space after , or .
    whole_txts.at[index,'fixed']=s #together they make sure there is at least one space between punc.
    index+=1
```

```python
whole_txts.head(2)
```

```python
list(whole_txts)
```

```python
#drop all other columns so it's just the text, don't have to do this
# just a preference

whole_=whole_txts.drop(columns=['Unnamed: 0','index','anon_id','native_language',
 'answer_id','question_id','text_len','text','level_id','processed_text',
 'wordtype_len'])
whole_.head(2)
```

```python
#writes to the WHOLE folder, the index is the name of the file 0.txt etc
i=0
for i, row in whole_.iterrows():
    if i>len(whole_):
        break
    else:
        f=open('C:\\Users\\Hanna\\Documents\\all_texts\\WHOLE\\'+str(i)+'.txt','w')
        f.write(whole_.at[i,'fixed'])
        f.close()  
        i+=1
```
