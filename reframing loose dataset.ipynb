```python
#importing libraries
import pandas as pd
```

```python
df=pd.read_csv('CSV\loose.csv')
df34=df.query('level_id==3 or level_id==4').reset_index().drop(columns=['index','Unnamed: 0'])
df45=df.query('level_id==5 or level_id==4').reset_index().drop(columns=['index','Unnamed: 0'])
```

```python
def get_uniques(x):
    return list(x.unique())
```

```python
# this code gets how many times an id appeared (n_anon_ids) as well as an array of what levels they participated in (lvl_nums)
clean_ids34=df34.groupby('anon_id').agg(
    n_anon_ids=('level_id', 'nunique'),
    lvl_nums=('level_id', get_uniques)
)

#changes the clean_ids_lvl's index so we can access the anon_id easier
clean_ids34.reset_index()
clean_ids34
```

```python
df34.query('anon_id=="ha2"')
```

```python
prog34=clean_ids34.query('n_anon_ids==2').reset_index()
prog34
```

```python
len(prog34)
```

```python
x=prog34['anon_id'].tolist()
```

```python
x
```

```python
df34
```

```python
ind=0
for i in df34.anon_id:
    if i not in x:
        df34.drop(index=ind,inplace=True)
    ind+=1
df34
```

```python
# this code gets how many times an id appeared (n_anon_ids) as well as an array of what levels they participated in (lvl_nums)
q_ids34=df34.groupby('anon_id').agg(
    n_anon_ids=('question_id', 'nunique'),
    lvl_nums=('level_id', get_uniques),
    q_nums=('question_id',get_uniques)
)

#changes the clean_ids_lvl's index so we can access the anon_id easier
q_ids34.reset_index(inplace=True)
q_ids34
```

```python
qgroup_ids34=df34.groupby('question_id').agg(
    n_anon_ids=('anon_id', 'nunique'),
    anon_ids=('anon_id',get_uniques)
)

#changes the clean_ids_lvl's index so we can access the anon_id easier
qgroup_ids34.reset_index(inplace=True)
qgroup_ids34
```

```python
q34=qgroup_ids34.query('n_anon_ids>1')
```
