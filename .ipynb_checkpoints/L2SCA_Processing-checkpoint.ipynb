```python
#importing libraries
import os
import numpy as np
import pandas as pd
```

```python
whole=pd.read_csv("CSV/whole_no_duplicates.csv")
loose=pd.read_csv("CSV/loose_no_duplicates.csv")
```

```python
#loose SCA metrics
l0_500=pd.read_csv('SCA_METRICS\loose_0-500.csv')
l501_1000=pd.read_csv('SCA_METRICS\loose501-1000.csv')
lend=pd.read_csv('SCA_METRICS\loose-1001-1807.csv')

#whole
w0_500=pd.read_csv('SCA_METRICS\whole0-500.csv')
w501_1000=pd.read_csv('SCA_METRICS\whole501-1000.csv')
w1001_1500=pd.read_csv('SCA_METRICS\whole1001-1500.csv')
w1501_2000=pd.read_csv('SCA_METRICS\whole1501-2000.csv')
w2001_2500=pd.read_csv('SCA_METRICS\whole2001-2500.csv')
w2501_3000=pd.read_csv('SCA_METRICS\whole2501-3000.csv')
w3001_3500=pd.read_csv('SCA_METRICS\whole3001-3500.csv')
w3501_4000=pd.read_csv('SCA_METRICS\whole3501-4000.csv')
w4001_4500=pd.read_csv('SCA_METRICS\whole4001-4500.csv')
w4501_5000=pd.read_csv('SCA_METRICS\whole4501-5000.csv')
w5001_5500=pd.read_csv('SCA_METRICS\whole5001-5500.csv')
w5501_6000=pd.read_csv('SCA_METRICS\whole5501-6000.csv')
w6001_6676=pd.read_csv('SCA_METRICS\whole6001-6676.csv')
```

```python
#combining loose to one dataset
loose_metric=l0_500.append(l501_1000, ignore_index=True).append(lend, ignore_index=True)
loose_metric
```

```python
#combining whole to one dataset
whole_metric=w0_500.append(w501_1000, ignore_index=True).append(w1001_1500, ignore_index=True).append(
w1501_2000, ignore_index=True).append(w2001_2500, ignore_index=True).append(w2501_3000, ignore_index=True).append(
    w3001_3500, ignore_index=True).append(w3501_4000, ignore_index=True).append(w4001_4500, ignore_index=True).append(
    w4501_5000, ignore_index=True).append(w5001_5500, ignore_index=True).append(w5501_6000, ignore_index=True).append(w6001_6676, ignore_index=True)

whole_metric
```

# Merging with existing datasets

```python
#creating name for txt files to merge on
whole['text_file']=pd.NaT
index=0
for text in whole.index:
    num=str(index)
    file_name=num+'.txt'
    whole.at[index,'text_file']=file_name
    index+=1
    
loose['text_file']=pd.NaT
index=0
for text in loose.index:
    num=str(index)
    file_name=num+'.txt'
    loose.at[index,'text_file']=file_name
    index+=1
```

```python
loose
```

```python
loose_metric.rename(columns={'Unnamed: 0':'text_file'},inplace=True)
whole_metric.rename(columns={'Unnamed: 0':'text_file'},inplace=True)
```

```python
loose_merged=loose.merge(loose_metric, on='text_file')
whole_merged=whole.merge(whole_metric, on='text_file')
```

```python
loose_merged.drop(columns={'Unnamed: 0','index'},inplace=True)
loose_merged.reset_index(inplace=True)
```

```python
whole_merged.drop(columns={'Unnamed: 0','index'},inplace=True)
whole_merged.reset_index(inplace=True)
```

```python
whole_merged
```

```python
loose_merged
```

```python
loose_merged['ratio']=pd.NaT
index=0
for i in loose_merged.index:
    unique=loose_merged.at[index,'wordtype_len']
    length=loose_merged.at[index,'text3_len']
    ratio=unique/length
    loose_merged.at[index,'ratio']=ratio
    index+=1
```

```python
whole_merged['ratio']=pd.NaT
index=0
for i in whole_merged.index:
    unique=whole_merged.at[index,'wordtype_len']
    length=whole_merged.at[index,'text_len']
    ratio=unique/length
    whole_merged.at[index,'ratio']=ratio
    index+=1
```

```python
def get_uniques(x):
    return list(x.unique())
```

```python
lvls=loose_merged.groupby('anon_id').agg(
    n_anon_ids=('level_id','nunique'),
    lvl_nums=('level_id',get_uniques)
)

ind_loose=lvls.reset_index()
ind_loose
```
